```python
import pandas as pd
import numpy as np
from scipy import stats
df_07 = pd.read_csv('/datasets/project_sql_result_07.csv')
df_07.head()
```

```python
# Convertir la columna 'start_ts' a formato datetime
df_07['start_ts'] = pd.to_datetime(df_07['start_ts'])
# Filtrar solo los sábados (0 = lunes, 6 = domingo)
df_07['day_of_week'] = df_07['start_ts'].dt.dayofweek
df_saturdays = df_07[df_07['day_of_week'] == 5]  # 5 corresponde a sábado
# Filtrar los sábados lluviosos (usando 'Bad' en el nombre de las condiciones)
saturdays_rainy = df_saturdays[df_saturdays['weather_conditions'].str.contains('Bad', case=False, na=False)]

# Filtrar los sábados no lluviosos (usando 'Good' en el nombre de las condiciones)
saturdays_no_rain = df_saturdays[df_saturdays['weather_conditions'].str.contains('Good', case=False, na=False)]

# Imprimir los resultados
print(saturdays_rainy)
print(saturdays_no_rain)
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

Bien hecho!<b/>

La manera de separar la información de ambos días es correcta
</div>

```python
# Eliminar valores nulos en 'duration_seconds'
saturdays_rainy = saturdays_rainy.dropna(subset=['duration_seconds'])
saturdays_no_rain = saturdays_no_rain.dropna(subset=['duration_seconds'])

# Realizar la prueba t
t_stat, p_value = stats.ttest_ind(saturdays_rainy['duration_seconds'], saturdays_no_rain['duration_seconds'])

# Mostrar los resultados
print(f'Estadístico t: {t_stat}')
print(f'Valor p: {p_value}')
```

```python
# Comparar el valor p con el nivel de significación (α = 0.05)
alpha = 0.05
if p_value <= alpha:
    print("Rechazamos la hipótesis nula: Hay una diferencia significativa en la duración de los viajes entre los sábados lluviosos y no lluviosos.")
else:
    print("No rechazamos la hipótesis nula: No hay una diferencia significativa en la duración de los viajes entre los sábados lluviosos y no lluviosos.")
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

Bien hecho!<br/>

Es correcto rechazar la hipotesis nula, debido a que sí hay notables cambios entre un día de buenas condiciones y uno con malas condiciones
</div>

*La hipótesis nula: la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare no cambia los sábados lluviosos.
*La hipótesis alternativa: La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos.
*Utilice un criterio de decisión de prueba t para muestras independientes, donde si el valor p ≤ 0.05, se rechaza la hipótesis nula.
* Dado que el valor p es pequeño, podemos rechazar la hipótesis nula con un alto nivel de confianza. Esto indica que sí hay una diferencia significativa en la duración de los viajes entre los sábados lluviosos y no lluviosos.
